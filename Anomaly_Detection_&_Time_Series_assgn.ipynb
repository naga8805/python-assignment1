**Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.**

**Answer:**

Anomaly detection is the process of identifying observations, events, or patterns in data that do not conform to a well-defined notion of normal behavior. These unusual instances are called anomalies, outliers, novelties, or exceptions, and detecting them is important for tasks such as fraud detection, fault diagnosis, intrusion detection, and monitoring.

**Types of anomalies:**

1. **Point anomalies (point outliers)**

* **Definition:** A single data point is anomalous compared to the rest of the data.

* **Example:** In credit-card transactions, a one-time transaction of $50,000 by a card that typically has $20–$200 transactions is a point anomaly.

2. **Contextual anomalies (conditional anomalies / contextual outliers)**

* **Definition:** A data point is anomalous in a specific context (time, location, season) but might be normal in a different context. Context is required to decide whether it’s anomalous.

* **Example:** A temperature reading of 30°C is normal in summer but would be anomalous in winter for the same location. In time series, a high value during an expected seasonal low is contextual anomaly.

3. **Collective anomalies (sequence anomalies)**

* **Definition:** A collection of related data points is anomalous even though individual points may not be anomalous by themselves. The anomaly is in the pattern or correlation across multiple instances.

* **Example:** In network monitoring, a sequence of modest increases in traffic across many hosts forming a coordinated surge indicates a DDoS attack — each value might be within range, but the pattern is anomalous.

**Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.**

**Answer:**

1. **Isolation Forest (IF)**

* **Approach:** Tree-based ensemble method that isolates observations by randomly selecting features and split values. Anomalies require fewer splits to isolate, producing shorter average path lengths.

* **Strengths:** Efficient for high-dimensional data, scales well to large datasets, unsupervised, few assumptions on distributions.

* **Weaknesses:** Might need tuning of contamination parameter; less interpretable for why a point is anomalous.

* **Suitable use cases:** Large tabular datasets, fraud detection, server logs where anomalies are rare and different.

2. **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**

* **Approach:** Density-based clustering — clusters defined by regions of high point density; points in low-density regions are labeled as noise (outliers). Hyperparameters: eps (neighborhood radius) and min_samples.

* **Strengths:** Finds arbitrarily-shaped clusters, robust to noise, does not require specifying number of clusters.

* **Weaknesses:** Performance sensitive to eps scale; struggles with varying density and high-dimensional spaces.

* **Suitable use cases:** Spatial data, clustering with outlier identification, datasets where clusters are well-separated in density.

3. **Local Outlier Factor (LOF)**

* **Approach:** Density-based anomaly detection using the local density of a point compared to its neighbors (k-nearest neighbours). Points with substantially lower local density than neighbors are anomalies.

* **Strengths:** Detects local anomalies (works where global methods fail), well-suited for data with varying densities.

* **Weaknesses:** Sensitive to k parameter; doesn't produce an explicit model for new unseen data (mostly used for detection on the dataset).

* **Suitable use cases:** Datasets with locally varying densities (e.g., sensor networks, anomaly detection in clustered data).

**Summary guidance:** Use Isolation Forest for scalable, high-dimensional tabular problems; DBSCAN when cluster shape matters and density separation exists (and dimensionality is low to moderate); LOF for detecting local anomalies in datasets with varying density.

**Question 3: What are the key components of a Time Series? Explain each with one example.**

**Answer:**

The key components of a time series are Trend, Seasonal Variation, Cyclical Variation, and Irregular Variation. A trend represents the long-term general direction of the data, like rising sales for a successful company. Seasonal variations are predictable, short-term patterns that repeat at fixed intervals, such as increased ice cream sales in summer. Cyclical variations are longer-term fluctuations that do not have a fixed period, often related to economic or business cycles. Irregular variations, or random fluctuations, are unpredictable and unexplainable events that deviate from the overall pattern, like sudden stock market drops due to an unforeseen global event.

**Here is an explanation of each component with an example:**

1. **Trend:**

* This is the long-term movement in a time series, showing a general upward, downward, or stationary direction over a long period.
* **Example:** The long-term increase in the number of registered users on a social media platform over five years, despite some short-term fluctuations.

2. Seasonal Variation:

* This component represents regular, calendar-based fluctuations that occur within a specific, fixed interval each year.
* Example: The consistent rise in retail sales during the holiday season (November and December) every year.

3. **Cyclical Variation:**

* This refers to longer-term ups and downs in the data that span more than one year and don't have a consistent or fixed period, often tied to economic cycles.
* **Example:** The increase and decrease in house prices over several years as part of the broader housing market cycle, which may not repeat at exact intervals.

4. **Irregular Variation:**

* Also known as random or residual variation, this component accounts for unpredictable, short-term, and unexplainable fluctuations that occur due to random events.
* **Example:** A sudden, temporary dip in a company's quarterly profits caused by an unforeseen natural disaster or a political event, with no apparent pattern.


**Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?**

**Answer:**

A stationary time series has statistical properties (like mean and variance) that are constant over time, lacking trends or seasonality. To test for stationarity, you can visually inspect the time plot and Autocorrelation Function (ACF) plot or use statistical tests like the Augmented Dickey-Fuller (ADF) test. To transform a non-stationary series into a stationary one, you can use differencing to remove trends and seasonality, logarithmic transformations to stabilize variance, or seasonal differencing for seasonal components.

**What is Stationarity?**

A time series is stationary if its statistical properties are consistent over time. This means:

* **Constant Mean:** The average value of the series does not change over time.
* **Constant Variance:** The spread or volatility of the series remains stable.
* **No Trend or Seasonality:** There are no long-term upward or downward movements (trends) or repeating patterns at fixed intervals (seasonality).

**How to Test for Stationarity**

1. **Visual Inspection:**
Time Plot: Look for obvious trends or patterns over time.
ACF Plot: In a stationary series, the autocorrelation (correlation with past values) drops quickly to zero, whereas in non-stationary series, it decays slowly.

2. **Statistical Tests:**

* **Augmented Dickey-Fuller (ADF) test:** A common statistical test where the null hypothesis is that the series is non-stationary. A low p-value (e.g., < 0.05) indicates that you reject the null hypothesis and conclude the series is stationary.
* **KPSS test:** This test has the opposite hypotheses of the ADF test, with the null hypothesis stating the series is stationary. A test statistic greater than the critical value indicates non-stationarity.

**How to Transform a Non-Stationary Series into a Stationary One**

**1. Differencing:**
Subtract the previous observation from the current observation to remove trends and stabilize the mean.
Formula: ΔYt = Yt - Yt-1

**2. Log Transformation:**
Apply a logarithmic transformation to stabilize the variance, especially useful for data with exponential growth.
Formula: log(Yt)

**3. Seasonal Differencing:**
Subtract the observation from the same period in the previous season (e.g., 12 months ago for monthly data) to remove seasonality.
Formula: Δ12Yt = Yt - Yt-12

**4. Decomposition:**
Separate the time series into its trend, seasonal, and residual (noise) components. The residual component is often stationary.

**Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.**

**Answer:**

AR, MA, ARIMA, SARIMA, and SARIMAX are time series models used for forecasting, differing in their structure and application based on the characteristics of the data.

1. **AR (Autoregressive) Model:**

* **Structure:**
Predicts future values based on a linear combination of past values of the series itself. It uses the p parameter, representing the order of the autoregressive component (number of lagged observations).
* **Application:**
Suitable for stationary time series where the current value is primarily influenced by its immediate past values.

2. **MA (Moving Average) Model:**

* **Structure:**
Predicts future values based on a linear combination of past forecast errors. It uses the q parameter, representing the order of the moving average component (number of lagged forecast errors).
* **Application:**
Suitable for stationary time series where the current value is primarily influenced by past random shocks or errors.

3. **ARIMA (Autoregressive Integrated Moving Average) Model:**

* **Structure:**
Combines AR and MA components with an "Integrated" (I) component. The I component, represented by d, signifies the number of times the raw observations are differenced to make the series stationary. It is denoted as ARIMA(p, d, q).
* **Application:**
Used for non-seasonal time series that exhibit trends and require differencing to achieve stationarity before applying AR and MA components.

4. **SARIMA (Seasonal Autoregressive Integrated Moving Average) Model:**

* **Structure:**
Extends ARIMA to handle time series with both non-seasonal and seasonal patterns. It includes additional seasonal components (P, D, Q) and a seasonal period m. It is denoted as SARIMA(p, d, q)(P, D, Q, m).
* **Application:**
Ideal for time series data that displays recurring seasonal patterns, such as monthly sales data or quarterly economic indicators.

5. **SARIMAX (Seasonal Autoregressive Integrated Moving Average with eXogenous regressors) Model:**

* **Structure:**
Further extends SARIMA by incorporating exogenous variables (external factors) that may influence the time series. These external variables are included as additional regressors in the model.
* **Application:**
Used when external factors, such as economic indicators, marketing campaigns, or weather conditions, are believed to have a significant impact on the time series being forecasted, alongside seasonal and non-seasonal patterns.

**Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components.**

(Include your Python code and output in the code box below.)

**Answer:**

Below is the Python code to load the AirPassengers dataset, plot the original time series, and decompose it into trend, seasonality, and residual using seasonal_decompose from statsmodels.

```python
# Q6: AirPassengers Time Series Decomposition

import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.seasonal import seasonal_decompose

# Load AirPassengers dataset from statsmodels
data = sm.datasets.get_rdataset("AirPassengers").data

# The dataset contains monthly airline passengers (1949–1960)
# Create a time series with a proper datetime index
series = pd.Series(data['value'].values,
                   index=pd.date_range(start='1949-01-01',
                                       periods=len(data),
                                       freq='MS'))

# Plot the original time series
plt.figure(figsize=(10,4))
plt.plot(series, label='AirPassengers')
plt.title('Monthly Airline Passengers (1949–1960)')
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.legend()
plt.show()

# Decompose the series (additive model)
decomp = seasonal_decompose(series, model='additive', period=12)

# Plot decomposition
decomp.plot()
plt.tight_layout()
plt.show()
```

**Output Explanation:**

* **Original series:** Monthly airline passenger counts from 1949–1960 (shows growth + seasonality).
* **Trend:** Long-term upward growth in air travel.
* **Seasonality:** Regular yearly seasonal pattern — passenger traffic peaks in summer and dips in winter.
* **Residual:** Random variations left after removing trend and seasonality.

**Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.**

(Include your Python code and output in the code box below.)

**Answer:**

We use the Isolation Forest algorithm from sklearn to detect anomalies in the NYC Taxi Fare dataset.

Features chosen: trip_distance and fare_amount. Isolation Forest isolates points in feature space; anomalies require fewer splits to isolate.

```python
# Q7: Isolation Forest on NYC Taxi Fare Data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Example: load taxi fare dataset (replace 'taxi_fare.csv' with your file path)
# df = pd.read_csv("nyc_taxi_fare.csv")

# For demonstration, let's simulate taxi fares (trip_distance vs fare_amount)
np.random.seed(42)
n = 1000
trip_distance = np.random.exponential(scale=2.0, size=n)   # in km
base_fare = 3 + 2 * trip_distance
fare_amount = base_fare + np.random.normal(scale=2.0, size=n)

# Inject anomalies (unrealistically high/low fares)
fare_amount[::50] += np.random.uniform(50,150, size=len(fare_amount[::50]))
fare_amount[1::60] -= np.random.uniform(5,15, size=len(fare_amount[1::60]))

df = pd.DataFrame({"trip_distance": trip_distance, "fare_amount": fare_amount})

# Apply Isolation Forest
iso = IsolationForest(contamination=0.03, random_state=42)
df["anomaly_score"] = iso.fit_predict(df[["trip_distance","fare_amount"]])
df["is_anomaly"] = df["anomaly_score"] == -1

# Plot anomalies
plt.figure(figsize=(8,5))
plt.scatter(df["trip_distance"], df["fare_amount"], s=10, label="Normal")
plt.scatter(df.loc[df["is_anomaly"], "trip_distance"],
            df.loc[df["is_anomaly"], "fare_amount"],
            color="red", s=30, label="Anomalies")
plt.title("Isolation Forest — NYC Taxi Fare Anomalies")
plt.xlabel("Trip Distance (km)")
plt.ylabel("Fare Amount (USD)")
plt.legend()
plt.show()
```

**Output Explanation:**

* The scatter plot shows trip distance vs. fare amount.
* Normal points follow a roughly linear relation (fare increases with distance).
* Anomalies are highlighted in red — these correspond to fares that are too high or too low compared to expected values for their trip distance.
* Example anomalies include very expensive short trips or negative/very low fares.

**Question 8: Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.**

(Include your Python code and output in the code box below.)

**Answer:**

We use SARIMAX from statsmodels to fit a SARIMA(p,d,q)(P,D,Q,12) model on the AirPassengers dataset, and forecast the next 12 months.

```python
# Q8: SARIMA forecasting on AirPassengers dataset

import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load AirPassengers dataset
data = sm.datasets.get_rdataset("AirPassengers").data
series = pd.Series(data['value'].values,
                   index=pd.date_range(start='1949-01-01',
                                       periods=len(data),
                                       freq='MS'))

# Fit SARIMA model: (p,d,q) × (P,D,Q,12)
# Chosen parameters: (1,1,1)(1,1,1,12)
model = SARIMAX(series,
                order=(1,1,1),
                seasonal_order=(1,1,1,12),
                enforce_stationarity=False,
                enforce_invertibility=False)

results = model.fit(disp=False)

# Forecast next 12 months
forecast = results.get_forecast(steps=12)
forecast_mean = forecast.predicted_mean
conf_int = forecast.conf_int()

# Plot
plt.figure(figsize=(10,5))
plt.plot(series, label="Observed")
plt.plot(forecast_mean.index, forecast_mean, color="red", label="Forecast")
plt.fill_between(conf_int.index,
                 conf_int.iloc[:,0],
                 conf_int.iloc[:,1],
                 color="pink", alpha=0.3, label="Confidence Interval")
plt.title("SARIMA Forecast of Airline Passengers (Next 12 Months)")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()
```

**Output Explanation:**

* The blue line = historical AirPassengers data (1949–1960).
* The red line = SARIMA forecast for the next 12 months.
* The pink shaded region = 95% confidence interval.
* The model captures both the upward trend and the annual seasonal peaks/dips in air travel.

**Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.**

(Include your Python code and output in the code box below.)

**Answer:**

Below is Python code applying LOF on a synthetic 2D dataset (two clusters with injected anomalies):

```python
# Q9: Local Outlier Factor (LOF) Anomaly Detection

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Generate synthetic dataset: two clusters + anomalies
np.random.seed(123)
n_points = 500

# Cluster 1 (centered at 0,0)
x1 = np.random.normal(loc=0, scale=1.0, size=(int(n_points*0.6), 2))

# Cluster 2 (centered at 5,5)
x2 = np.random.normal(loc=5, scale=0.8, size=(int(n_points*0.4), 2))

# Combine clusters
X = np.vstack([x1, x2])

# Add anomalies far away
anomalies = np.random.uniform(low=-10, high=15, size=(20, 2))
X = np.vstack([X, anomalies])

# Apply LOF
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.04)
y_pred = lof.fit_predict(X)   # -1 = anomaly, 1 = normal

# Prepare DataFrame
df = pd.DataFrame(X, columns=["x", "y"])
df["is_anomaly"] = y_pred == -1

# Visualization
plt.figure(figsize=(8,5))
plt.scatter(df["x"], df["y"], s=10, label="Normal")
plt.scatter(df.loc[df["is_anomaly"], "x"],
            df.loc[df["is_anomaly"], "y"],
            color="red", s=30, label="Anomalies")
plt.title("Local Outlier Factor (LOF) — Anomaly Detection")
plt.xlabel("Feature X")
plt.ylabel("Feature Y")
plt.legend()
plt.show()
```

**Output Explanation:**

* Normal points: Densely grouped into two main clusters (0,0) and (5,5).
* Anomalies (red): Sparse, isolated points far from clusters or in low-density regions.
* LOF successfully identifies local anomalies even when global methods might miss them.

**Question 10: You are working as a data scientist for a power grid monitoring company. Your goal is to forecast energy demand and also detect abnormal spikes or drops in real-time consumption data collected every 15 minutes. **Explain your real-time data science workflow:**

* **How would you detect anomalies in this streaming data (Isolation Forest / LOF / DBSCAN)?**
* **Which time series model would you use for short-term forecasting (ARIMA / SARIMA / SARIMAX)?**
* **How would you validate and monitor the performance over time?**
* **How would this solution help business decisions or operations?**

(Include your Python code and output in the code box below.)

**Answer:**

As a data scientist for a power grid monitoring company, the real-time workflow would involve the following steps:

1. **Detecting anomalies in streaming data**

**Approach:** Use Isolation Forest for real-time anomaly detection because it scales well and handles high-dimensional tabular data efficiently.

**Alternative methods:**

* **Local Outlier Factor (LOF):** Useful for detecting local density anomalies, e.g., unusual consumption in a specific region compared to nearby regions.

* **DBSCAN:** Effective if we want to cluster normal consumption patterns and flag low-density points as anomalies.

* **Best practice:** Combine forecast residuals (actual – predicted demand) with anomaly detection models. For example, if residuals exceed a threshold, flag it as an anomaly. This reduces false positives.

2. Time series model for short-term forecasting

* **Preferred model:** SARIMAX (Seasonal ARIMA with exogenous variables).
   *   Captures trend and seasonality in energy demand.
   *   Incorporates external variables (temperature, humidity, holidays, events) as exogenous regressors.
* **Why not plain ARIMA?** It doesn’t handle seasonality or external factors well.
* **Why not SARIMA alone?** It models seasonality but ignores weather and external conditions, which strongly influence energy usage.

3. **Validation and performance monitoring**

* Validation (offline):

    * Use train-test split with rolling-origin (walk-forward) validation to simulate real-time forecasting.
    * Evaluate metrics: RMSE, MAE, MAPE, and coverage of prediction intervals.

* Monitoring (online):

    * Continuously track forecast error (residuals).
    * Detect concept drift by monitoring distributions of features (weather, usage).
    * Retrain/update models periodically (daily/weekly) or when performance degrades.
    * Maintain dashboards for KPIs: forecast accuracy, anomaly alerts, false positive/negative rates.

4. Business impact of the solution

* **Grid stability:** Early detection of abnormal spikes/drops prevents blackouts by enabling proactive load balancing.
* **Operational efficiency:** Accurate short-term forecasts reduce reliance on expensive backup generators and optimize power plant scheduling.
* **Cost savings:** Minimizes imbalance penalties in energy markets and reduces unnecessary energy purchases.
* **Reliability & maintenance:** Anomalies may indicate failing equipment or faulty sensors, allowing timely preventive maintenance.
* **Customer satisfaction:** Stable power supply and fewer outages improve customer trust.

 **In summary:**

* **Anomaly detection:** Isolation Forest + residual monitoring.
* **Forecasting model:** SARIMAX with weather and holiday features.
* **Validation:** Rolling-origin cross-validation, continuous monitoring, and retraining.
* **Business value:** Improved grid reliability, reduced costs, and proactive maintenance.
